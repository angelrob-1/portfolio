```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
# clean up data
# only want USA races, 50k or 50mi, 2020
```

```python
# Step 1 show 50km or 50mi
```

```python
df[df['Event distance/length'] == '50km']
```

```python
# combine 50km/50mi
```

```python
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# combine all filters
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
usa_races = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
usa_races.head(10)
```

```python
# remove USA from event name
```

```python
usa_races['Event name'].str.split('(').str.get(0)
```

```python
usa_races['Event name'] = usa_races['Event name'].str.split('(').str.get(0)
```

```python
usa_races.head(10)
```

```python
# clean up athlete age
```

```python
usa_races['athlete age'] = 2020 - usa_races['Athlete year of birth']
```

```python
# remove h from athlete performance
```

```python
usa_races['Athlete performance'] = usa_races['Athlete performance'].str.split(' ').str.get(0)
```

```python
usa_races.head(5)
```

```python
# drop columns: Athlete club, Athlete country, Athlete year of birth, Athlete age category
```

```python
usa_races = usa_races.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category' ], axis = 1)
```

```python
usa_races.head()
```

```python
# clean up null values
```

```python
usa_races.isna().sum()
```

```python
usa_races[usa_races['athlete age'].isna() == 1]
```

```python
usa_races = usa_races.dropna()
```

```python
usa_races.shape
```

```python
# check for duplicates
```

```python
usa_races[usa_races.duplicated() == True]
```

```python
# reset index
```

```python
usa_races.reset_index(drop = True)
```

```python
# fix data types
```

```python
usa_races.dtypes
```

```python
usa_races['athlete age'] = usa_races['athlete age'].astype(int)
```

```python
usa_races['Athlete average speed'] = usa_races['Athlete average speed'].astype(float)
```

```python
usa_races.dtypes
```

```python
usa_races.head()
```

```python
# rename columns
```

```python
for col in usa_races.columns:
    print(col)
```

```python
usa_races = usa_races.rename(columns = {'Year of event':'year', 
                                        'Event dates': 'race_date',
                                        'Event name': 'race_name',
                                        'Event distance/length': 'race_length',
                                        'Event number of finishers': 'race_number_of_finishers',
                                        'Athlete performance': 'athlete_performance',
                                        'Athlete gender': 'athlete_gender',
                                        'Athlete average speed': 'athlete_average_speed',
                                        'Athlete ID': 'athlete_id',
                                        'athlete age': 'athlete_age' })
```

```python
usa_races.head()
```

```python
# reorder columns
```

```python
new_usa_races = usa_races[['race_date', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 
                           'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
new_usa_races
```

```python
sns.histplot(new_usa_races['race_length'])
```

```python
sns.histplot(new_usa_races, x= 'race_length', hue = 'athlete_gender')
```

```python
sns.displot(new_usa_races[new_usa_races['race_length'] == '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(new_usa_races, x = 'race_length', y = 'athlete_average_speed', hue = 'athlete_gender')
```

```python
# Insights from data
```

```python
# Difference in speed for 50km,50mi male to female
```
